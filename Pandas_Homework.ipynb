```python
## import depedendcies
import pandas as pd
import os
import numpy as np
```

```python
# Import data

school_data = 'Pandas_Homework\\py_city_schools.csv'
student_data = 'Pandas_Homework\\py_city_students.csv'
```

```python
# Create DF's

school_data_df = pd.read_csv(school_data, index_col = "School ID")
student_data_df = pd.read_csv(student_data, index_col = "Student ID")
```

```python
# merge data

combined = school_data_df.merge(student_data_df, how = "outer", on = "school_name")
 
 
```

```python





combined.head()
```

```python
# District Summary

Total_Schools = len(combined["school_name"].unique())
Total_Students = len(combined["student_name"])
Total_Budget = combined["budget"].unique().sum()
Avg_Math = round(combined["math_score"].sum()/len(combined["math_score"]),2)
Avg_Read = round(combined["reading_score"].sum()/len(combined["reading_score"]),2)
# going to assume the 70 and above is passing since no guidelines given. 
mask1 = combined["math_score"] >= 70. 
Math_Percent_Pass = round(len(combined[mask1]["math_score"])/len(combined["math_score"]) * 100,2)

mask2 = combined["reading_score"] >= 70  #/len(combined["reading_score"])
Reading_Percent_Pass = round(len(combined[mask2]["reading_score"])/len(combined["reading_score"]) * 100,2)
Overall_Passing = round((Math_Percent_Pass + Reading_Percent_Pass)/2,2)
```

```python
Summary = [Total_Schools, Total_Students, Total_Budget,Avg_Math,Avg_Read,
                            Math_Percent_Pass,Reading_Percent_Pass,Overall_Passing]
labels = ["Total Number of Schools","Total Number of Students","Total District Budget",
          "Average Math Score","Average Reading Score","Percent Passing Math","Percent Passing Reading", 
          "Overall Percent Passing"] 
District_Summary = pd.Series(Summary, index =  labels )
```

```python
District_Summary
```

```python
# School Summary
combined["passing_math"] = mask1
combined["passing_reading"] = mask2

schools = combined.groupby("school_name")



schooltyp = schools["type"].max()
stutot = schools["student_name"].count()
schoolnm = schools["school_name"].max().values
psb = schools["budget"].mean().values
perstub = round(schools["budget"].mean()/schools["budget"].count(),2)

avr = round(schools["reading_score"].mean(),2)
avm = round(schools["math_score"].mean(),2)
ppm = round(schools["passing_math"].sum()/stutot* 100,2)
ppr = round(schools["passing_reading"].sum()/stutot*100,2)
pptot = (ppr+ppm)/2

school_dict = {"School Name": schoolnm,"Type": schooltyp, "Size": stutot, 
               "School Budget": psb,
               "Per Student Budget": perstub,"Average Reading Score": avr, "Average Math Score": avm,
               "Percent Passing Reading": ppr, "Percent Passing Math": ppm, 
               "Total Percent Passing": pptot}



Per_School_Summary = pd.DataFrame(school_dict)
Per_School_Summary.set_index("School Name",inplace = True)
Per_School_Summary

```

```python

# By Budget/Student


bins = [575,590,605,620,635,660]
group_div= ["$575 - 589","$590 - 604","$605 - 619","$620 - 634","$635 - 659" ]




Per_School_Summary["By Budget"]= pd.cut(Per_School_Summary["Per Student Budget"],bins, labels = group_div)
School_Breakdown = Per_School_Summary[["By Budget","Average Math Score","Average Reading Score","Percent Passing Math",
                  "Percent Passing Reading","Total Percent Passing"]]
School_Breakdown.set_index("By Budget", inplace = True)
breakdown_grp = School_Breakdown.groupby("By Budget")



```

```python
breakdown_grp.mean()
```

```python
# by School Size


bins = [0,1000,2000,5000]
group_div= ["Small (<1000)","Medium(1000 - 2000)","Large - (2000 - 5000)" ]

Per_School_Summary["By Size"]= pd.cut(Per_School_Summary["Size"],bins, labels = group_div)
School_Breakdown = Per_School_Summary[["By Size","Average Math Score","Average Reading Score","Percent Passing Math",
                  "Percent Passing Reading","Total Percent Passing"]]
School_Breakdown.set_index("By Size", inplace = True)
breakdown_grp_2 = School_Breakdown.groupby("By Size")
breakdown_grp_2.mean()

# # this makes it pretty clear that the large schools struggle against medium and small in effectiveness.
## also of the 8 charters schools only one of them falls in the large category 
```

```python
#By School Type
School_Breakdown = Per_School_Summary[["Type","Average Math Score","Average Reading Score","Percent Passing Math",
                  "Percent Passing Reading","Total Percent Passing"]]

breakdown_grp_3 = School_Breakdown.groupby("Type")
breakdown_grp_3.mean()
```
